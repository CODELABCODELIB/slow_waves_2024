```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_increasing_frequency_spikes(duration, base_freq, freq_increase_rate, amplitude, noise_level):
    # Time vector
    t = np.linspace(0, duration, int(duration * 1000))  # 1000 samples/second

    # Frequency as a function of time
    freq = base_freq + freq_increase_rate * t

    # Phase as the integral of frequency
    phase = np.cumsum(freq) * 2 * np.pi

    # Signal with spikes
    signal = amplitude * np.sin(phase) + noise_level * np.random.randn(len(t))

    return t, signal

# Example usage
duration = 10  # seconds
base_freq = 1  # Hz
freq_increase_rate = 0.5  # Hz/s
amplitude = 2
noise_level = 0

t, signal = simulate_increasing_frequency_spikes(duration, base_freq, freq_increase_rate, amplitude, noise_level)

plt.plot(t, signal)
plt.xlabel("Time (s)")
plt.ylabel("Signal")
plt.title("Simulating Increasing Frequency Spikes")
plt.show()
```
